```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

using pandas import and read excel file of canada immegrants 

```python
df_can = pd.read_excel( "canada.xlsx",                           #file name/ file path / file url
                         sheet_name = "Canada by Citizenship",  #getting data from sheet
                          skiprows = 20,   #skipping unwanted 20 rows from original datasets 
                          skipfooter = 2)  #skipping footer from original dataset

print("data read into panda dataframe")

df_can.head(6) #read first six rows from the table
```

```python
#let's convert dataframe column and index to List using .to_list() method
column_list = df_can.columns.to_list()
index_list = df_can.index.to_list()
print(column_list,"\n",index_list)
```

```python
'''now let's clean the data by removing unnecessory columns from the tables and ranaming the necessory one's by usig
df.drop() and df.rename() methods
 '''
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis = 1 , inplace = True )
df_can.head(2)
```

```python
#axis=1 specifies that we are working with df.columns and axis=0 means df.rows
df_can.rename({'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, axis=1,inplace = True)
print(df_can.columns)
df_can.head(2)
```

```python
#we added a new column total which sums up the total immigrants
df_can['Total'] = df_can.sum(axis=1)
df_can.head(2)
```
